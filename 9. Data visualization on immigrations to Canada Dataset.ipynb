### Get started with dataset

```python
import numpy as np
import pandas as pd

df = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                  sheet_name = 'Canada by Citizenship', skiprows = range(20), skipfooter = 2)
df.head()
```

```python
print(type(df.columns))
print(type(df.index))
```

```python
df.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis = 1, inplace = True)
df.head()
```

```python
df.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace = True)
df.head()
```

```python
df['Total'] = df.sum(axis = 1)
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# years = list(map(str, range(1980, 2014)))
India = df.loc["India", range(1980, 2014)]
India.index = India.index.map(int)
India.plot(kind="line")
plt.title('Immigration from India')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
df_1 = df.loc[['India', 'China'], range(1980, 2014)]
df_1.head()
```

```python
df_1 = df_1.transpose()
df_1.head()
```

```python
df_1.plot(kind = "line")
plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
df.sort_values(by = "Total", ascending = False, axis = 0, inplace = True)

df_1 = df.head()
df_1
```

```python
df_1 = df_1[range(1980, 2014)].transpose()
df_1
```

```python
df_1.plot(kind="line", figsize = (12, 12))
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```
