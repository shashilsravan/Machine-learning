### Area plots

```python
import numpy as np
import pandas as pd
df = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2
                      )

print('Data downloaded and read into a dataframe!')
```

```python
df.head()
```

```python
df.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)

df.head()
```

```python
df.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'}, inplace=True)

df.head()
```

```python
all(isinstance(column, str) for column in df.columns)
```

```python
df.columns = list(map(str, df.columns))
all(isinstance(column, str) for column in df.columns)
```

```python
df.set_index('Country', inplace = True)
df.head()
```

```python
df['total'] = df.sum(axis = 1)
df.head()
```

```python
years = list(map(str, range(1980, 2014)))
years
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
```

```python
df.sort_values(['total'], ascending = False, axis = 0, inplace = True)
df.head()
```

```python
df_1 = df.head()
df_1 = df_1[years].transpose()
df_1.head()
```

```python
df_1.plot(kind = 'area', figsize = (12, 12), stacked = False)

plt.title("Immigration Trend of top 5 countries")
plt.xlabel('Years')
plt.ylabel("Number of immigrants")
plt.show()
```

```python
df_1.plot(kind = 'area', figsize = (12, 12), stacked = True)
plt.title("Immigration Trend of top 5 countries")
plt.xlabel('Years')
plt.ylabel("Number of immigrants")
plt.show()
```

```python
df_1.plot(kind='area', alpha=0.25, stacked=False, figsize=(10, 10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
ax = df_1.plot(kind='area', alpha=0.35, figsize=(10, 10))

ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

```python
df_2 = df.tail()
df_2 = df_2[years].transpose()
df_2.head()
```

```python
df_2.plot(kind='area', alpha = 0.45, figsize = (12, 12))
plt.title('Immigration Trend of 5 Countries with Least Contribution to Immigration')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
ax = df_2.plot(kind = "area", alpha = 0.35, stacked = False, figsize = (12, 12))
ax.set_title('Immigration Trend of 5 Countries with Least Contribution to Immigration')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

### Histograms

```python
df['2013'].head()
```

```python
count, bin_edges = np.histogram(df['2013'])
print(count)
print(bin_edges)
```

```python
df['2013'].plot(kind="hist", figsize=(8, 5))

plt.title('Histogram of Immigration from 195 Countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
count, bin_edges = np.histogram(df['2013'])

df['2013'].plot(kind='hist', figsize=(8, 5), xticks = bin_edges)

plt.title('Histogram of Immigration from 195 countries in 2013') 
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
df.loc[['Denmark', 'Norway', 'Sweden'], years].plot.hist()
```

```python
df.loc[['Denmark', 'Norway', 'Sweden'], years].transpose().plot(kind = "hist", figsize=(10, 6))
plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
df_t = df.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
count, bin_edges = np.histogram(df_t, 15)
df_t.plot(kind ='hist', figsize=(10, 6), bins=15, alpha=0.6, xticks=bin_edges, 
          color=['coral', 'darkslateblue', 'mediumseagreen'])

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
count, bin_edges = np.histogram(df_t, 15)
xmin = bin_edges[0] - 10   
xmax = bin_edges[-1] + 10  


df_t.plot(kind='hist', figsize=(10, 6), bins=15, xticks = bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen'], stacked=True, xlim=(xmin, xmax))

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants') 

plt.show()
```

```python
df_t = df.loc[["Greece", "Albania", "Bulgaria"], years]
df_t = df_t.transpose()

count, bin_edges = np.histogram(df_t, 15)

df_t.plot(kind = "hist", figsize = (10, 6), bins = 15, alpha = 0.35, xticks = bin_edges, color = ['coral', 'darkslateblue', 'mediumseagreen'])
plt.title('Histogram of Immigration from Greece, Albania, and Bulgaria from 1980 - 2013')
plt.ylabel("Number of years")
plt.xlabel("Number of immigrations")
plt.show()
```

```python
df_t = df.loc[["India"], years]
df_t = df_t.transpose()

count, bin_edges = np.histogram(df_t, 15)

df_t.plot(kind = "hist", figsize = (10, 6), bins = 15, alpha = 0.35, xticks = bin_edges, color = ['coral', 'darkslateblue', 'mediumseagreen'])
plt.title('Histogram of Immigration from India during 1980 - 2013')
plt.ylabel("Number of years")
plt.xlabel("Number of immigrations")
plt.show()
```

### Barplots

```python
df_i = df.loc["India", years]
df_i.head()
```

```python
df_i.plot(kind = "bar", figsize = (10, 6))

plt.xlabel('Year')
plt.ylabel('Number of immigrants') 
plt.title('Indian immigrants to Canada from 1980 to 2013') 
plt.show()
```

```python
df_ic = df.loc['Iceland', years]
df_ic.plot(kind = "bar", figsize = (10, 6), rot = 90)

plt.xlabel('Year')
plt.ylabel('Number of immigrants') 
plt.title('Indian immigrants to Canada from 1980 to 2013') 
plt.annotate('', xy=(32, 70),xytext=(28, 20),xycoords='data', 
            arrowprops = dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2))
plt.annotate('2008 - 2011 Financial Crisis', xy=(28, 30), rotation=72.5,va='bottom',ha='left')
plt.show()
```

```python
df.sort_values(by='total', ascending= True, inplace=True)
df_top15 = df['total'].tail(15)
df_top15.to_frame()
```

```python
df_top15.plot(kind = 'barh', figsize = (8, 8), color = 'steelblue')
plt.xlabel('Number of Immigrants')
plt.title('Top 15 Conuntries Contributing to the Immigration to Canada between 1980 - 2013')
plt.ylabel('Countries')
```

```python
for index, value in enumerate(df_top15):
    print("Index is {0} and the values is {1}".format(index, value))
```

```python
df_top15.plot(kind='barh', figsize=(12, 12), color='steelblue')
plt.xlabel('Number of Immigrants')
plt.title('Top 15 Conuntries Contributing to the Immigration to Canada between 1980 - 2013')

for index, value in enumerate(df_top15):
    label = format(int(value), '-')
    plt.annotate(label, xy = (value - 47000, index - 0.1), color = "white")
plt.show()
```
